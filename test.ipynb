```python
def get_subset(features_index, mutual_information, threshold=0.53):
    merged = zip(features_index, mutual_information)
    sorted_features = sorted(merged, key=lambda x: x[1], reverse=True)

    filtered_features = filter(lambda x: x[1] > threshold, sorted_features)
    return filtered_features


# filter is a generator so we must run a loop to check
for i in get_subset(
    features_index=[1,0,0], 
    mutual_information=[3,7,8],
    threshold=3
):
    print(i)
```
